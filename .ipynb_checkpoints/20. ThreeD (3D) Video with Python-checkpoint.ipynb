In this file, we will go through an excellent technique to create a 3D video with python by using the matplotlib package.

First, weâ€™ll create a simple 3D contour plot with matplotlib:

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create data points
x = np.linspace(-10, 10, 100)
y = np.linspace(-15, 15, 100)
```

```python
# create grid
X, Y = np.meshgrid(x, y)
Z = np.sin(X) + np.cos(Y)
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(9, 6))
ax = plt.axes(projection = '3d')

# hide the axis
ax._axis3don = False

# 3d contour plot
ax.contour3D(X, Y, Z, 100, cmap = 'viridis')

# make panel color white
ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))

# make grid color white
ax.xaxis._axinfo['grid']['color'] = (1, 1, 1, 0)
ax.yaxis._axinfo['grid']['color'] = (1, 1, 1, 0)
ax.zaxis._axinfo['grid']['color'] = (1, 1, 1, 0)

# adjust point of view
ax.view_init(60, 90)
```

The code above will give us a digit without an axis because we are declaring `axis3don = False`. We cannot hide the panel and grid colour, but we can make it similar to the background image `(white)` as mentioned in the comment `# make the panel and grid colour "white"`. 

A video is a collection of images. We are therefore going to create many images with different azimuthal axis viewpoints. We can adjust this:

```python
# adjust point of view
ax.view_init(60, 90)
```

Now, we need to change the value of `90` to a sequence number from `0` to `360` by using following code:

```python
plt.figure.max_open_warning # Figures created through the pyplot interface (`matplotlib.pyplot.figure`) are retained until explicitly closed and may consume too much memory. 
# (To control this warning, see the rcParam `figure.max_open_warning`).

for angle in range(0,360,2):
    plt.figure(figsize=(16, 9))
    ax = plt.axes(projection = '3d')
    ax._axis3don = False
    # 3d contour plot
    ax.contour3D(X, Y, Z, 200, cmap = 'viridis')
    ax.xaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
    ax.yaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
    ax.zaxis.set_pane_color((1.0, 1.0, 1.0, 0.0))
    ax.xaxis._axinfo['grid']['color'] = (1, 1, 1, 0)
    ax.yaxis._axinfo['grid']['color'] = (1, 1, 1, 0)
    ax.zaxis._axinfo['grid']['color'] = (1, 1, 1, 0)
    
    # adjust view from 0, 2, 4, 6, ..., 360
    ax.view_init(60, angle)
    
    # save figure with different names depend on the view
    filename='3d_video/3d_vis_'+str(angle)+'.png'
    plt.savefig(filename, dpi=75)
```

We will get 180 images with different names. We can check it in Jupyter notebook by using the following command:

```python
# ! ls 3d_video/
```

The names of the images are listed from `3d_vis_0.png` to `3d_vis_358_.png`. To combine them all, we need to create an array consisting of the names of figures, using the code below:

```python
# using pillow
from PIL import Image 

png_count = 180
files = []
for i in range(png_count):
    seq = str(i*2)
    file_names = '3d_vis_'+ seq +'.png'
    files.append(file_names)
    
print(files)
```

Now we will combine all of that into a single video to create a 3D video. The extension of the video is in **GIF**. We can combine it using follpwing code:

```python
# Create the frames
frames = []
for i in files:
    new_frame = Image.open("3d_video/"+i)
    frames.append(new_frame)
```

```python
# Save into a GIF file that loops forever   
frames[0].save('3d_video/3d_vis.gif', format='GIF',
               append_images=frames[1:],
               save_all=True,
               duration=40, loop=0)
```

It will save to a **GIF** file named `3d_vis.gif` that is 40 seconds long in a single loop.
